```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as py
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.templates
import seaborn as sns
from scipy import stats
from scipy.stats import norm,skew
```

```python
from sklearn.datasets import fetch_california_housing
fetch_california_housing=fetch_california_housing()
x=fetch_california_housing.data  #independent features
y=fetch_california_housing.target #dependent variable
data=pd.DataFrame(x,columns=fetch_california_housing.feature_names)
data['Price']=y
data.head()
```

```python
#finding the relationship between attributes
f,a=py.subplots()
a.scatter(x=data['MedInc'],y=data['Price'])
py.xlabel("MedInc")
py.ylabel("Price")
py.show()
```

```python
sns.distplot(data['Price'],fit=None)
(mean,sigma)=norm.fit(data['Price'])
print("mean: ",mean)
print("sigma: ",sigma)
py.plot()
fig=py.figure() #QQ plot
r=stats.probplot(data['Price'],plot=py)
py.show()
```

```python
data['Price']=np.log1p(data["Price"])
sns.distplot(data["Price"],fit=norm)
(mean,sigma)=norm.fit(data['Price'])
print("mean: ",mean)
print("sigma: ",sigma)
py.plot()
fig=py.figure() #QQ plot
r=stats.probplot(data['Price'],plot=py)
py.show()
```

```python
from sklearn.model_selection import train_test_split
x=data.drop("Price",axis=1)
y=data['Price']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
predictions=lr.predict(x_test)
print("Actual Value of the House: ",y_test[0])
print("Model predicted value of the House: ",predictions[0])
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predictions)
rmse=np.sqrt(mse)
print(mse)
print(rmse)
```
